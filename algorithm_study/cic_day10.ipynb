# 그래프

- BFS(너비 우선 탐색) : 먼저 들어온 노드가 먼저 나가야함(큐 자료구조 활용, 선입선출)
  - 최단 거리 탐색에 효과적
  - 많은 메모리 필요 -> 오버 플로우 발생 가능
- DFS(깊이 우선 탐색) : (스택 자료구조 활용, 후입선출, 재귀함수 활용, 재귀함수도 마지막에 들어온게 출력)
  - 경로를 탐색하는데 효과적
  - 주로 재귀함수 사용? <- 백트래킹에 효과적
  - 하나의 경로만 계속 저장

````python
import sys

sys.stdin = open("input.txt")
input = sys.stdin.readline

V = int(input())
E = int(input())

# 인접 리스트 그래프
graph_1 = []
for _ in range(V) : 
    temp = []
    graph_1.append(temp)

# 간선 정보 입력
for _ in range(E) :
    start, end = list(map(int, input().split()))
    graph_1[start].append(end)

for row in graph_1 :
    print(row)
````

````python
# 재귀 함수를 활용한 dfs
def dfs(node) :
    # bfs라면, node = queue.popleft()
    # 인접 노드를 탐색
    for adj_node in graph_1[node]:
        # 인접노트 방문확인
        if visited[adj_node] == False :
            # 방문한적이 없으면 인접 노드에 대해 방문 처리
            visited[adj_node] = True
            # 스택에 인접 노드를 저장
            # bfs라면, queue.append(node)
            # 스택을 활용했다면 stack.append(node)
            dfs(adj_node)

# 방문 확인 변수 필요
visited = [False] * V
# visited = [False,False, False, False, False, False, False]
# 0번 노드 방문 확인 visited[0] == False -> 방문을 안했구나

# 리스트 in 시간복잡도가 O(n) : 탐색 -> 비효율적
# 리스트의 인덱싱 [index] 시간복잡도 O(1) : 접근

# 시작 노드 처리
start = 0
visited[start] = True # 시작 노드는 방문 처리

# DFS를 시작 = DFS 호출
dfs(start)

# DFS는 스택메모리 활용
````

````python
# 놓쳤다

def dfs_2(node) :
    for adj_node in graph_1[node]:
        # 방문 확인
        if visited[adj_node] != True :
            # 인접 노드 방문 처리
            visited[adj_node] = True
            dfs_2(adj_node)
            

# 시작노드
start = 0

# 방문 변수
visited = [False] * V

visited[start] = True

# DFS 시작, 스택에 시작 노드를 저장
# == DFS 함수 호출
dfs_2(start)
````

```
문제
그래프를 DFS로 탐색한 결과와 BFS로 탐색한 결과를 출력하는 프로그램을 작성하시오. 단, 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문하고, 더 이상 방문할 수 있는 점이 없는 경우 종료한다. 정점 번호는 1번부터 N번까지이다.

입력
첫째 줄에 정점의 개수 N(1 ≤ N ≤ 1,000), 간선의 개수 M(1 ≤ M ≤ 10,000), 탐색을 시작할 정점의 번호 V가 주어진다. 다음 M개의 줄에는 간선이 연결하는 두 정점의 번호가 주어진다. 어떤 두 정점 사이에 여러 개의 간선이 있을 수 있다. 입력으로 주어지는 간선은 양방향이다.
```

````python
import sys
sys.stdin = open('input.txt')

import sys
input = sys.stdin.readline
N, M, V = list(map(int, input().strip().split()))

graph = []

# N + 1 만큼 빈 리스트 추가

for _ in range(N+1):
    temp = []
    graph.append(temp)

# M개의 줄 간선의 정보
for _ in range(M):
    start, end = list(map(int, input().strip().split()))
    graph[start].append(end)
    graph[end].append(start)

for index in range(len(graph)) :
    graph[index] = sorted(graph[index])

for row in graph :
    print(row)
````

````python
import sys
sys.stdin = open('input.txt')

import sys
input = sys.stdin.readline

N, M, V = list(map(int, input().strip().split()))

# 인접 리스트 표현
graph = []

# N + 1만큼 빈 리스트 추가
for _ in range(N + 1):
    temp = []
    graph.append(temp)

# M개의 줄 간선의 정보
for _ in range(M):
    # 양방향 간선
    start, end = list(map(int, input().strip().split()))
    graph[start].append(end)
    graph[end].append(start)

for index in range(len(graph)):
    graph[index] = sorted(graph[index])


# DFS 구현

stack = list()
start = V

# 방문 처리 변수
visited = set()

# 시작 정점에 대한 방문 처리, 큐 저장
visited.add(start)
stack.append(start)

# ----------------------------------- 여기까지 dfs를 위한 재료 준비

while stack :
    # 탐색의 기준이 되는 노드를 꺼내기
    node = stack.pop() # 리스트 마지막에 있는 값을 꺼내옴
    print(node, end = " ") # -> 거쳐간 노드의 경로 확인 가능
    # 인접 노드를 순회

    for adj_node in graph[node]:
        # 인접 노드 방문 확인
        if adj_node not in visited :
            # 탐색 노드 저장
            # 방문 처리
            stack.append(adj_node)
            visited.add(adj_node)

# DFS와 BFS 출력을 구분하는 줄바꿈 코드
print()

# BFS 구현

from collections import deque

queue = deque()
start = V

# 방문 처리 변수
visited = set()

# 시작 정점에 대한 방문 처리, 큐 저장
visited.add(start)
queue.append(start)

# ----------------------------------- 여기까지 bfs를 위한 재료 준비

while queue :
    # 탐색의 기준이 되는 노드를 꺼내기
    node = queue.popleft()
    print(node, end = " ") # -> 거쳐간 노드의 경로 확인 가능
    # 인접 노드를 순회

    for adj_node in graph[node]:
        # 인접 노드 방문 확인
        if adj_node not in visited :
            # 탐색 노드 저장
            # 방문 처리
            queue.append(adj_node)
            visited.add(adj_node)
````

````python
# 백준 2606 바이러스
# https://www.acmicpc.net/problem/2606

import sys
sys.stdin = open('input.txt')

import sys
input = sys.stdin.readline
# 첫 번째 정점의 수
V = int(input())
E = int(input())

# 인접 리스트 표현
graph = []
for _ in range(V+1): # 정점의 번호가 1번부터 시작
    graph.append([])

# 간선의 정보 E개
for _ in range(E):
    start, end = list(map(int, input().split()))
    graph[start].append(end)
    graph[end].append(start)

# for row in graph :
#     print(row)

start = 1
count = 0

stack = []
stack.append(start)
visited = set()
visited.add(start)

while stack :
    node = stack.pop()
    # print(node)
    count += 1
    
    #인접 노드 순회
    for adj_node in graph[node]:
        #인접 노드 방문 확인
        if adj_node not in visited:
            visited.add(adj_node)
            stack.append(adj_node)

print(count - 1)
````
