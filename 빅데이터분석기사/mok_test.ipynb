```python
# Min-Max 정규화
Min = df['Wind'].min()
Max = df['Wind'].max()
df['min_max'] = round((df['Wind']-Min)/(Max-Min), 2)

# z정규화
Mean = df['Wind'].mean()
Std = df['Wind'].std()
# 표준 편차(standard deviation)
df['z'] = round((df['Wind']-Mean)/Std, 2)

a = df['min_max'].mean()
b = df['z'].mean()

print(a)
print(b)
```

# 제2유형

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('Bank_Personal_Loan_Modeling.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

- 결측치 제거
- 의미없는 변수 제거

```python
ploan_processed = df.dropna().drop(['ID', 'ZIP Code'], axis=1, inplace = False)
ploan_processed.head()
```

- 독립, 종속 변수 분리

```python
X = ploan_processed.drop(['Personal Loan'], axis = 1)
y = ploan_processed['Personal Loan']
```

- 전처리
    - train, test, valid 나누기 
    - 전처리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=y,train_size = 0.7, test_size = 0.3, random_state=1234)
# stratify=y : 타겟 변수 y의 클래스 비율을 유지하며 데이터를 나눈다.

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
import sklearn.preprocessing as preprocessing

preprocessor = preprocessing.Normalizer()
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
# 모델 생성

from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

# 1에서 25까지 n_neighbors를 적용 
neighbors_settings = range(1,25)

# KNeighborsClassifier() 함수에 K개수 입력 -> 모델 만들고 데이터 적용 
for n_neighbors in neighbors_settings :

    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)

    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))

    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

```
